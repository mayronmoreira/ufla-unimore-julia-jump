## Capacitated Vehicle Routing Problem (CVRP) 

The Capacitated Vehicle Routing Problem (CVRP) is one of the most challenging combinatorial optimization problems. Its prominence in real contexts has called the attention of companies and shorten the gap between enterprises and academia. In this notebook, we deal with the two-index flow formulation of the CVRP, and we extend this model to include time windows constraints. As follows, we present a Figure extracted from [The blog of Logistics at MGEPS at UPV](https://logisticsmgepsupv.wordpress.com/2014/04/10/vehicle-routing-problem-its-there-a-perfect-solution/comment-page-1/), that represents three routes and one depot.

### CVRP: mathematical formulation

Let $C=\{2,...,n\}$ be a set of customers. A demand $q_i \in \mathbb{R}_+$ is associated with each customer $i \in C$. Consider a homogeneous fleet composed of $K$ vehicles with a maximum capacity of $Q$. Each vehicle must start and end its route in the depot. All customers must be visited exactly one.  We model the input data using a complete graph $G=(N,E)$, where $N = C \cup \{0,n+1\}$ is the set of nodes representing customers and the depot nodes $1$ and $n+1$ (note that $q_1 = q_{n+1} = 0$). The idea is to state that each route must start at node $1$ and finish at node $n+1$ (which represent, in fact, the same place). The cost of travel from node $i$ to node $j$ is given by $w_{ij} \in \mathbb{R}_+$, where $(i,j) \in E$. 

**Goal**: to minimize the route costs, ensuring that all the problem requirements are satisfied.

* **Parameters**

    * $C=\{2,...,n\}$: set of customers;
    * $N = C \cup \{1,n+1\}$: set of nodes;
    * $q_i \in \mathbb{R}_+$: demand of node $i \in N$;
    * $Q$: capacity of the vehicle;
    * $K$: number of vehicles;
    * $w_{ij} \in \mathbb{R}_+$: travel distance between nodes $i$ and $j$.

#### Mathematical model

The mathematical model was adapted from [Munari et al. (2016)](https://arxiv.org/abs/1606.01935). (*Write it with time limit of 30 seconds, with 1 thread*)

* **Variables**

    * $x_{ij} \in \{0,1\}$: equal to one if there is a route from customer $i$ to $j$ directly. $i,j \in N$;
    * $y_{j} \in \mathbb{R}_+$: measures the cumulated demand on the route that visits node $j \in N$ up to this visit.

* **Objective function**: minimize travel costs.

\begin{equation}
\min \sum_{i=1}^{n+1}\sum_{j=1}^{n+1}w_{ij}x_{ij}
\end{equation}

* **Constraints 1**: each customer must be visited.

\begin{equation}
 \sum_{j=2, j \neq i}^{n+1}x_{ij} = 1 \quad i=2,...,n
\end{equation}

* **Constraints 2**: control the flow of vehicles through the graph.

\begin{equation}
 \sum_{i=1, i \neq h}^{n}x_{ih} - \sum_{j=2, j \neq h}^{n+1}x_{hj} = 0 \quad h=2,...,n
\end{equation}

* **Constraints 3**: limit the number of vehicles used.


\begin{equation}
 \sum_{j=2}^{n}x_{1j} \le K
\end{equation}

* **Constraints 4** and **Constraints 5**: guarantee that the capacity of vehicles is not surpassed. 

\begin{equation}
 y_j \ge y_i + q_jx_{ij} - Q(1-x_{ij}) \quad i,j=1,...,n+1
\end{equation}

\begin{equation}
 q_i \le y_i \le Q \quad i=1,...,n+1
\end{equation}

### Time-Windows Constraints

Time-windows constraints are useful since in practice customers impose a time interval to be attended. Formally, we include to the previous set of parameters an interval $[e_i,l_i]$ which establishes that the service at node $i \in C$ cannot start earlier than $e_i$ not later than $l_i$. We also define $s_i$ and $t_{ij}$ as the service time of node $i$ and travel time from node $i$ to node $j$.

```julia
# We generate random the additional data
s = [rand(1:30) for _ in 1:n+1]
s[1] = s[n+1] = 0

t = w # To simplify, we suppose that distance and travel matrixes are proportional

TW=Vector{Tuple{Int,Int}}() # Time Windows vector
tpl = (0,round(Int, maximum(w)))
push!(TW, tpl)
for i=2:n
    a = rand(1:30)
    b = a + s[i] + rand(1:30)
    push!(TW, (a,b))
end
push!(TW, tpl)
;
```

For this purpose, let $\alpha_i$ be a continuous variable that measures the instant that service time starts at node $i$. Then, we include in the previous model the following constraints:

\begin{equation}
\alpha_j \ge \alpha_i + (s_i + t_{ij})x_{ij} - M_{ij}(1-x_{ij}) \quad i=1,...,n, j=2,...,n+1
\end{equation}

where $M_{ij} = \max_{i,j=1,...,n+1}\{l_i - e_j,0\}$.

```julia
# Definition of constant M
M = zeros(Int, n+1, n+1)
for i=1:n+1, j=1:n+1
    M[i,j] = max(TW[i][2]-TW[j][1],0)
end

println("Write your code here!")
```

### Another comments about Vehicle Routing Formulations

One of the most known formulations for the CVRP is the set partitioning model. In this case, we suppose that we have a set of routes $\mathcal{R}$ that satisfy the problem requirements. Then we create a binary variable $\lambda_r$ that is equal to one if the route $r \in \mathcal{R}$ is selected to be part of the final solution. Taking $c_r$ as the cost of the route $r$ and $a_{ri}$ as a binary parameter equal to one if the route $r$ visits the customer $i$, the model reads as follows.

\begin{equation}
\min \sum_{r \in \mathcal{R}} c_r\lambda_r
\end{equation}

subject to:

\begin{equation}
\sum_{r \in \mathcal{R}} a_{ri}\lambda_r = 1 \quad \forall i \in C
\end{equation}

\begin{equation}
\sum_{r \in \mathcal{R}} \lambda_r \le K
\end{equation}

\begin{equation}
\lambda_r \in \{0,1\} \quad \forall r \in \mathcal{R}
\end{equation}

The *column generation* is widely applied to this formulation. In Julia, we suggest the new framework called [*Coluna.jl*](https://github.com/atoptima/Coluna.jl). A presentation about this package is shown at https://www.juliaopt.org/meetings/santiago2019/slides/vitor_nesello.pdf.

To implement *Benders Decomposition*, we suggest https://github.com/mlubin/JuliaBenders.
